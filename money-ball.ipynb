```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
%matplotlib inline
```

```python
player_data = pd.read_csv('data/Master.csv')
batting = pd.read_csv('data/BattingPost.csv')
salaries = pd.read_csv('data/Salaries.csv')
positions = pd.read_csv('data/Appearances.csv')
fielding
salaries.head()
```

```python
positions.head()
```

```python
batting['OBP'] = batting[['H','BB', 'HBP']].sum(1)/batting[['AB', 'BB', 'HBP', 'SF']].sum(1)
```

```python
#combining play, batting, salaries, and position data.
df = pd.merge(pd.merge(player_data,batting), salaries)
df = pd.merge(df,positions)
```

```python
#remove dead players
df_alive = df[df.deathYear.isnull()]
df_alive[df_alive['yearID'] == 2015].head()
```

```python
#what columns are currently avialable?
print(df_alive.columns)
```

```python
#filtering df alive
df_filtered = df_alive[['playerID', 'nameFirst', 'nameLast','salary','OBP', 'yearID', 'G', 'AB',
       'R', 'H', '2B', '3B', 'HR', 'RBI', 'SB', 'CS', 'BB', 'SO', 'IBB', 'HBP',
       'SH', 'SF', 'GIDP', 'G_all', 'GS', 'G_batting',
       'G_defense', 'G_p', 'G_c', 'G_1b', 'G_2b', 'G_3b', 'G_ss', 'G_lf',
       'G_cf', 'G_rf', 'G_of', 'G_dh', 'G_ph', 'G_pr']]
```

```python
#2015 data
df_filtered_2015 = df_filtered[df_filtered.yearID == 2015]
#
df_filtered_2015[df_filtered.HBP > 1]
```

```python
#Removing noise
temp = df_filtered[(df_filtered.OBP != 0)]
temp1 = temp[(temp.OBP.isnull() == False)]
temp2 = temp1[temp1.OBP != 1]
```
